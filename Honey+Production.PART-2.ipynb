```python
#3. Display the percentage distribution of the data in each year using the pie chart.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
per=df.groupby('year').size()/len(df)*100
```

```python
plt.figure(figsize=(5,5))
plt.pie(per,labels=per.index,autopct='%1.1f%%')
plt.show()
```

```python
#6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x='year', y='prodvalue', data=df)
```

Findings:
1.The boxplot displays the distribution of 'prodvalue' for each year.
2.It shows where the middle 50% of the data lies.
The line inside the box represents the median (middle) value of 'prodvalue' for each year.
3.This visualization will help you understand how 'prodvalue' is distributed across different years and whether there are any significant differences or trends over time.

```python
#8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'stocks', 'price per lb', 'prodvalue')
```

```python
sns.heatmap(df[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']].corr())
```

Inference:
1.The heatmap provides a visual representation of the correlation matrix, where each cell represents the correlation coefficient between two variables.
2.Correlation coefficients range from -1 (perfect negative correlation) to 1 (perfect positive correlation), with 0 indicating no linear correlation.
3.In this heatmap, positive correlations are shown in warmer colors, while negative correlations are in cooler colors.
